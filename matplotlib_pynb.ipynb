```python
import matplotlib.pyplot as plt
import numpy as np

#we are using labels to beautify our graphs a bit
x =np.array([2,4,6,8,10,12])
y = np.array([3,6,9,12,15,18])

plt.title('Experimental graph')
plt.xlabel('pulse')
plt.ylabel('calorie burnt')

plt.plot(x,y,color='hotpink',marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#this is gridded example

x = np.array([2,5,7,8,9,10,12])
y = np.array([2,4,6,8,10,13,14])

plt.title('Avg student life')
plt.xlabel('the more u study')
plt.ylabel('thre more u r going to score')

plt.plot(x,y)
plt.grid(color='green',lw=2,ls=':')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#this will show the colormao plus the scatter feature 
x = np.array([2,4,6,8,10,12,14,16])
y = np.array([3,4,5,7,8,9,13,15])
color = np.array([10,20,25,30,35,40,60,80])

plt.scatter(x,y,c= color,cmap='viridis')

plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#bar graph experimentation
x =np.array(['A','B','C','D'])
y =np.array([20,40,30,35])

plt.bar(x,y,width=0.5,color='hotpink')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#will do frequency distribution using plt random
x = np.random.normal(170,10,250)

plt.hist(x)
plt.show()
```
